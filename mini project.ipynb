```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('sales.csv')
df.head()
```

```python
print(df.isnull().sum())
df.fillna(0,inplace=True)
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['Total_Sales']=np.multiply(df['Units_Sold'],df['Unit_Price'])
df.head()
```

```python
total_revenue=df['Total_Sales'].sum()
avg_units=df.groupby('Product')['Units_Sold'].mean()
top_product=df.groupby('Product')['Total_Sales'].sum().idxmax()
print(f"Total Revenue:{total_revenue}")
print("\n Average Units Sold per product:\n",avg_units)
print(f"\nTop sellinf=g produuct:{top_product}")
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['Date'],df['Total_Sales'],marker='',color='black')
plt.title('Daily Sales trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
Product_sales = df.groupby('Product')['Total_Sales'].sum()
plt.figure(figsize=(6,4))
plt.bar(Product_sales.index,Product_sales.values,color=['hotpink','yellow','green'])
plt.title('Daily Sales trend')
plt.xlabel('Product')
plt.ylabel(' Sales Amount')
plt.show()
```
